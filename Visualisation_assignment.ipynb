```python
#Q5. . Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
#
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(123)

# Generate synthetic dataset with correlated features
n_samples = 1000
n_features = 5

# Define correlation structure
cov_matrix = np.array([
    [1.0, 0.8, 0.7, 0.4, 0.2],
    [0.8, 1.0, 0.9, 0.6, 0.3],
    [0.7, 0.9, 1.0, 0.8, 0.5],
    [0.4, 0.6, 0.8, 1.0, 0.7],
    [0.2, 0.3, 0.5, 0.7, 1.0]
])

# Generate multivariate normal data
data = np.random.multivariate_normal(mean=[0, 0, 0, 0, 0], cov=cov_matrix, size=n_samples)

# Create pandas DataFrame
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5'])

# Calculate correlation matrix
corr_matrix = df.corr()

# Visualize correlation matrix using heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()


```

```python
 #Q3. 4. Using the given x and y data, generate a 3D surface plot to visualize the function z= sin sqrt(x**2+y**2)
import plotly.graph_objects as go
import numpy as np
import pandas as pd
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
```

```python
df
```

```python
fig = go.Figure(data=[go.Surface(x=df.X, y=df.Y, z=df.Z)])
fig.update_layout(title='3D Surface Plot')

fig.show()
```

```python
#Q1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
!pip install bokeh
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show


# Create x-values from 0 to 10
x = np.linspace(0, 10, 100)

# Calculate y-values as sine of x
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, line_width=2, legend_label="sin(x)", line_color="blue")

# Show plot
show(p)
```

```python
#Q2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
#  markers based on the 'sizes' and 'colors' columns.
np.random.seed(0)
data = {
    'x': np.random.normal(0, 10, 100),
    'y': np.random.normal(0, 10, 100),
    'sizes': np.random.uniform(5, 20, 100),
    'colors': np.random.choice(['blue', 'red', 'green', 'yellow', 'purple'], 100)
}

# Create DataFrame
df = pd.DataFrame(data)

# Create Bokeh plot
p = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y')

# Add scatter renderer to plot
p.scatter(x='x', y='y', size='sizes', fill_color='colors', line_color='black', alpha=0.7, source=df)

# Show plot
show(p)
```

```python
#Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
data = {"Fruit" : ['Apples', 'Oranges', 'Bananas', 'Pears'],
"Count" : [20, 25, 30, 35]}
df= pd.DataFrame(data)

p = figure(title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count',
           x_range=df['Fruit'])

# Add bar renderer to plot
p.vbar(x='Fruit', top='Count', width=0.9, source=df, legend_field='Fruit',
       line_color='white', fill_color='navy')

# Show plot
show(p)

```

```python
#Q4. Create a Bokeh histogram to visualize the distribution of the given data.
!pip install bokeh
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

# Add histogram renderer to plot
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
      fill_color="navy", line_color="white", alpha=0.7)


show(p)
```
