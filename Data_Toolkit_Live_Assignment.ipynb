```python
#Q.3. Explain the following terms
#The difference in np.array, np.asarray and np.asanyarray

#Ans. np.array:
#      Creates a new array object, even if the input is already an array.
#      Copies the input data by default.
#      Allows specifying a dtype, which can lead to data conversion.
# 2. np.asarray:
#      Returns the input as an array, without copying if possible (e.g., if input is already an array).
#      Does not allow specifying a dtype, preserving the original data type.
# 3. np.asanyarray:
#      Similar to np.asarray, but returns an array of the same subclass (if input is a subclass of ndarray).


# The difference between Deep copy and shallow copy
#Ans. . Shallow Copy
#      Creates a new object, but references the same underlying data.
#      Changes to the copied object's data affect the original object.
#      np.array, np.asarray, and np.asanyarray create shallow copies by default.
# 2. Deep Copy:
#      Creates a new object and copies the underlying data.
#      Changes to the copied object's data do not affect the original object.
#      To create a deep copy, use np.copy()

```

```python
#Q14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
#Perform the following tasks using Matplotlib.

# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
# right y-axis for 'Humidity').

# b) Label the x-axis as 'Date'.

# c) Set the title of the plot as 'Temperature and Humidity Over Time'.
```

```python
# Create a time-series dataset in a Pandas DataFrame
date_range = pd.date_range('2024-01-01', periods=100)
temperature = np.random.uniform(20, 30, 100)
humidity = np.random.uniform(50, 90, 100)
df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})
df
```

```python
# Plot 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(df['Date'], df['Temperature'], color='red', marker='o')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='blue', marker='o')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

plt.title('Temperature and Humidity Over Time')
plt.grid(True)
plt.show()
```

```python
#Q17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
#origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

x = np.random.randn(100)
y = np.random.randn(100)

plt.figure(figsize=(8, 8))
sns.scatterplot(x=x, y=y, hue=np.sign(x) * np.sign(y), palette="coolwarm", legend="full")

plt.legend(title="Quadrant", loc="upper right", labels=["Q1 (x>0, y>0)", "Q2 (x<0, y>0)", "Q3 (x<0, y<0)", "Q4 (x>0, y<0)"])
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Quadrant-wise Scatter Plot")
plt.show()
```

```python
import numpy as np
from bokeh.plotting import figure, show

# Generate x values
x = np.linspace(0, 4*np.pi, 100)

# Generate sine wave values
y = np.sin(x)

# Create the line chart
p = figure(title='Sine Wave Function', x_axis_label='x', y_axis_label='sin(x)', height=350, width=600)

# Add grid lines
p.grid.grid_line_alpha = 0.3
p.grid.grid_line_color = "gray"

# Plot the sine wave
p.line(x, y, line_width=2, line_color="blue")

# Show the plot
show(p)
```
